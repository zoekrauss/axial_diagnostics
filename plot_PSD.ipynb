```python
from obspy import read
from obspy.io.xseed import Parser
from obspy.signal import PPSD
import obspy
import matplotlib.pyplot as plt
```

```python
st = read("AXAS1_20230713.mseed")
tr = st.select(id="OO.AXAS1..EHZ")[0]
```

```python
inv = obspy.core.inventory.inventory.read_inventory("OO.xml")
ppsd = PPSD(tr.stats, metadata=inv)
```

```python
ppsd.add(st)
```

```python
fig = ppsd.plot(show=False,show_noise_models=False);
ax = fig.axes
```

```python
fig.set_figwidth(15)
fig.set_figheight(8)
ax[0].set_xlim([0.01,100])
fig
```

```python
ppsd.plot(cumulative=True)
```
